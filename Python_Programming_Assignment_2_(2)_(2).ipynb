```python
#How does the survival rate vary by passenger class (pclass)?

sns.countplot(x='pclass' , hue='survived' , data=data)
plt.title ('Survival Rate by Passenger Class')
plt.show()
```

```python
#How does the survival rate vary by age (age)?

sns.histplot(data['age'],kde=True)
plt.title ('Distribution of Ages of Survival')
plt.show()
```

```python
#How does the survival rate vary by passenger sex (sex)?

sns.countplot(x='sex' , hue='survived' , data=data)
plt.title ('Survival Rate by Passenger Sex')
plt.show()
```

```python
#How does the survival rate vary by family size (sibsp, parch?

data['family_size'] = data['sibsp'] + data['parch'] + 1
sns.countplot(x='family_size', hue='survived' , data=data)
plt.title ('Survival Rate by Passenger Family Size')
plt.show()
```
